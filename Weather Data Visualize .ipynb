```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
bike = pd.read_csv("day.csv")
```

```python
bike.head()
```

## Scatter plot Corrlation

```python
bike.info()
```

```python
bike.dteday = pd.to_datetime(bike.dteday)
```

```python
plt.plot(bike.dteday , bike.temp)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.plot(bike.dteday , bike.windspeed)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.plot(bike.dteday , bike.cnt)
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.scatter(bike.cnt,bike.temp)
plt.show()
```

```python
bike.cnt.corr(bike.temp)
```
